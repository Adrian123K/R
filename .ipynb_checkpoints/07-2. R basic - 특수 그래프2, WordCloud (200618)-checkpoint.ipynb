## ■ WordCloud
    R로 WordCloud를 그리려면 R java 설치 필요

```r
library(plyr)
library(KoNLP)
library(wordcloud)
```

```r
useSejongDic() # 세종 사전에 있는 한글을 R로 로드
```

```r
winter<-readLines('winter.txt')
nouns<-sapply(winter, extractNoun, USE.NAMES = F)
nouns<-unlist(nouns)
nouns<-nouns[nchar(nouns)>=2]
nouns
```

```r
cnouns<-count(nouns)
cnouns
```

```r
pal<-brewer.pal(6,'Dark2')
pal<-pal[-(1)]
# 글씨 폰트 설정
windowsFonts(malgun=windowsFont("맑은 고딕"))
```

```r
wordcloud(words=cnouns$x,        # 단어
          freq=cnouns$freq,       # 단어 빈도수
          colors=pal,            # 색깔
          min.freq=3,            # 빈도수가 3개 이상인 것만 시각화
          random.order=F,        # F로 하게되면 큰 글씨부터 출력하여 중앙에서부터 
          family='malgun')       # 폰트
```
